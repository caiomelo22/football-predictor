```python
import warnings

from helpers import classification as pf
from helpers.file import save_file
from helpers.options import filtered_cols, selected_stats, strategy

warnings.filterwarnings("ignore")
```

```python
save_pipeline = True
league = "bundesliga"
seasons = "2017-2023"
season_test = 2023
betting_starts_after_n_games = 0
```

```python
# Read the data
X_full, y, X_test_full, y_test, odds_test = pf.get_league_data(
    league, seasons, season_test
)
```

```python
# Keep selected columns only
X_train, y_train, X_test = pf.filter_datasets(X_full, y, X_test_full, filtered_cols)
```

```python
my_pipeline = pf.simulate(
    X_train, y_train, X_test, y_test, odds_test, betting_starts_after_n_games, strategy, verbose=1
)
```

```python
cols_info = {
    "filtered_cols": filtered_cols,
    "selected_stats": selected_stats,
    "strategy": strategy,
}

# Save the pipeline
if save_pipeline:
    path = f"../dist/leagues/{league}/official"
    
    save_file(path, cols_info, 'columns')
    save_file(path, my_pipeline, 'pipeline')
```
